```python
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('classic')

def drawPoint(x, y):
  print(f'({x}, {y})')
  plt.scatter(x, y)
```

```python
def DDA(p1, p2): 
  x0, y0 = p1
  x1, y1 = p2
  x, y = x0, y0
  dx, dy = x1 - x0, y1 - y0  
  m = dy / dx if dx != 0 else 0
  pasos = abs(dy) if abs(dy) > abs(dx) else abs(dx)
  x_inc = dx / pasos if pasos != 0 else 0
  y_inc = dy / pasos if pasos != 0 else 1
  drawPoint(x, y)
  print('m = ', m)
  print('dx =', dx)
  print('dy =', dy)
  print('r =', pasos)
  print('delta_x', x_inc)
  print('delta_y', y_inc)
  if (x0 < 0 and x1 < 0):
    x_inc = x_inc + 0.1
  elif (x0 < 0 and x1 > 0):
    x_inc = x_inc - 0.1
  for _ in range(pasos): 
    x, y = x + x_inc, y + y_inc
    drawPoint(round(x), round(y))

# Test de derecha izquierda 

#DDA((1,1), (4, 4))
#DDA((4, 4), (1, 1))

#DDA((1,1), (4, 1))
#DDA((4, 1), (1, 1))

#DDA((1, 1), (1, 4))
#DDA((1, 4), (1, 1))

#DDA((1, 4), (4, 1))
#DDA((4, 1), (1, 4))


#DDA((2,3), (3,1))
#DDA((3, 1), (2, 3))

#DDA((-1, 1), (-2, -1))
#DDA((-2, -1), (-1, 1))

DDA((-1, 1), (0, -1))
plt.show()
```

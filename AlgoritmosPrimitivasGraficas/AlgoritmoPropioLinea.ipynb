```python
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('classic')

def drawPoint(x, y):
  print(f'({x}, {y})')
  plt.scatter(x, y)
```

```python
# Algoritmo completo para dibujar una lÃ­nea recta a partir dos puntos
# Basado en el algoritmo DDA
# Considera que la pendiente puede tomar valores positivos, negativos, cero o infinito
# Funciona de izquierda a derecha o de derecha a izquierda

def DDA(x0, y0, x1, y1):

  dx, dy = x1 - x0, y1 - y0
  m = dy / dx if dx != 0 else 0
  if m == 0: 
    if dx != 0:
      pasos, x_inc, y_inc = dx, 1, 0
    else:
      pasos, x_inc, y_inc = dy, 0, 1
  elif m >= 1:
    pasos, x_inc, y_inc = dx, 1, 1
  else:
    pasos, x_inc, y_inc = dy, 1, -1
  x, y = x0, y0
  drawPoint(x, y)
  if x0 > x1 or y0 > y1: 
      x_inc, y_inc = - x_inc, - y_inc
  for _ in range(abs(pasos)): 
    x, y = x + x_inc, y + y_inc
    drawPoint(round(x), round(y))

# De izquierda a derecha
#DDA(1, 1, 4, 4)
#DDA(1, 1, 4, 1)
#DDA(1, 1, 1, 4)
#DDA(1, 4, 4, 1)

# De derecha a izquierda
#DDA(1,4, 1, 1)
#DDA(4, 1, 1, 4)
#DDA(4, 1, 1, 1)
DDA(4, 4, 1, 1)
plt.show()
```

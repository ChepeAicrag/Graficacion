```python
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('classic')

def drawPoint(x, y):
  print(f'({x}, {y})')
  plt.scatter(x, y)
```

```python
def DDA(p1, p2): 
  x0, y0 = p1
  x1, y1 = p2
  flag = False 
  if x0 > x1 or (y0 > y1 and x0 == x1): 
      x_tmp,y_tmp = x0, y0
      x0, y0, x1, y1, flag = x1, y1, x_tmp, y_tmp, True
  dx, dy = x1 - x0, y1 - y0  
  x, y = x0, y0
  if flag: dx, dy, x, y = - dx, - dy, x1, y1
  m = dy / dx if dx != 0 else 0
  pasos = abs(dy) if abs(dy) > abs(dx) else abs(dx)
  x_inc = dx / pasos if pasos != 0 else 0
  y_inc = dy / pasos if pasos != 0 else 1
  drawPoint(x, y)
  print('m = ', m)
  print('dx =', dx)
  print('dy =', dy)
  print('r =', pasos)
  print('delta_x', x_inc)
  print('delta_y', y_inc)
  for _ in range(pasos): 
    x, y = x + x_inc, y + y_inc
    drawPoint(round(x), round(y))

# Test traslacion
#DDA((1,1), (4, 1))
DDA((1,4), (1, 1))
#DDA((4, 1), (1, 4))

#DDA((2,3), (3,1))
#DDA((3, 1), (2, 3))
plt.show()
```

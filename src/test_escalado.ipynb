```python
from Libreria.Functions import line, escalado
from Libreria.Utils import  draw_point, combinacion
```

```python
def generar_escalado(puntos = [], t = (0, 0)):
    points_direction = combinacion(puntos, 2)
    print (points_direction)
    pintar = []
    for i in points_direction:
        for j in line(i[0], i[1]):
            x, y = j
            xr, yr = escalado(p = j, pf = puntos[0], t = t)
            c, cr = 'blue', 'red'
            if (x, y, c) not in pintar and (x, y, cr) not in pintar:
                pintar.append((x, y, c))
            #if (xr, yr, cr) not in pintar and (xr, yr, c) not in pintar:
            pintar.append((xr, yr, cr))
            #draw_point(x, y, 'blue')
            #draw_point(xr, yr, 'red')  
    for i in pintar: 
        x, y, c = i
        draw_point(x, y, c)

if '__main__' == '__main__':
    p1 = (1, 1)
    p2 = (3, 1)
    p3 = (2, 3)
    T = (2, 3)
    generar_escalado(puntos = [p1, p2, p3], t = T)
```
